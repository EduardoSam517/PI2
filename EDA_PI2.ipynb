Se tiene como proposito hacer un mapa de calor con las columnas "pos x" y "pos y" pero por ahora las columnas sonde tipo object**

```python
tipo_pos_x = data["pos x"].dtype
tipo_pos_y = data["pos y"].dtype

print(f"Tipo de dato 'pos x': {tipo_pos_x}")
print(f"Tipo de dato 'pos y': {tipo_pos_y}")
```

```python
#EL siguiente codigo quita de ambas columnas todo lo que no tenga la estructura de un bool pero sin hacerlos bool, esto por que habia valores ".""

data["tipo_pos_x"] = data["pos x"].apply(lambda x: isinstance(x, float))
data["tipo_pos_y"] = data["pos y"].apply(lambda x: isinstance(x, float))

```

```python
data["pos x"] = pd.to_numeric(data["pos x"], errors="coerce")
data["pos y"] = pd.to_numeric(data["pos y"], errors="coerce")
```

```python
data = data.dropna(subset=["pos x", "pos y"])
data = data[data["pos x"].apply(lambda x: isinstance(x, float))]
data = data[data["pos y"].apply(lambda x: isinstance(x, float))]
```

```python
data["pos x"] = pd.to_numeric(data["pos x"])
data["pos y"] = pd.to_numeric(data["pos y"])
```

```python
# Crea una matriz de densidad a partir de las coordenadas x e y
density = sns.jointplot(x="Pos x", y="Pos y", data=data, kind="kde")
```

```python
# Importa la función kdeplot
import seaborn as sns

# Crea el mapa de calor
density = sns.jointplot(x="Pos x", y="Pos y", data=data, kind="kde")

# Ajusta la paleta de colores
density.plot_joint(palette="rocket", func=sns.kdeplot)

# Agrega un título y etiquetas a los ejes
plt.title("Mapa de calor de ubicaciones")
plt.xlabel("Posición X")
plt.ylabel("Posición Y")

# Ajusta el tamaño del mapa
plt.gcf().set_size_inches(10, 8)

# Muestra el mapa de calor
plt.show()

```

```python
import plotly.express as px
import numpy as np
from scipy.stats import pearsonr
```

```python
# Calcular la distancia entre cada par de puntos
distancias = np.sqrt(((data["Pos x"] - data["Pos x"].shift())**2) + ((data["Pos y"] - data["Pos y"].shift())**2))

# Calcular el coeficiente de correlación de Pearson
correlacion = pearsonr(data["Pos x"], data["Pos y"])[0]
```

```python
data_nuevo = data[["Pos x", "Pos y"]]
data_nuevo
```

```python
data_nuevo.isnull().sum()
```

```python
# Crear una cuadrícula de bins
x_bins = np.linspace(data_nuevo["Pos x"].min(), data_nuevo["Pos x"].max(), 20)
y_bins = np.linspace(data_nuevo["Pos y"].min(), data_nuevo["Pos y"].max(), 20)

# Calcular la densidad de puntos en cada bin
density = sns.histplot(data=data_nuevo, x="Pos x", y="Pos y", bins=[x_bins, y_bins], stat="density", 
                       palette="rocket", common_norm=False)

# Ajustar el título y las etiquetas de los ejes
plt.title("Mapa de calor de accidentes viales en Buenos Aires")
plt.xlabel("Posición X (Latitud)")
plt.ylabel("Posición Y (Longitud)")

# Mostrar el mapa de calor
plt.show()
```

```python
# Crear el mapa de calor
density = sns.histplot(data=data_nuevo, x="Pos x", y="Pos y", bins=[x_bins, y_bins], stat="density", 
                       palette="rocket", common_norm=False)

# Agregar una leyenda
plt.legend(title="Densidad de accidentes")

# Mostrar el mapa de calor
plt.show()
```
